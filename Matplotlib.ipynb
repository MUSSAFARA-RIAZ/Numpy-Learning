```python
#Matplotlib useful for making plots
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
print(x)
print("******************Sin value********************")
print(y)
print("*****************Cos value*******************")
print(z)
```

Plotting the data

```python
#sin wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title, x-axis and y-axis labels
plt.plot(x,y)
plt.xlabel("Angle")
plt.ylabel("Sin Value")
plt.title("Sine wave")
plt.show()
```

```python
# Parabola
x=np.linspace(-10,10,30)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'yd')
plt.show()
# y represent yellow color and + is used instead of line
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'r--')
plt.plot(x,np.cos(x),'g.')
plt.title("Graphs of Sin and Cosine func")
plt.show()
```

Bar Plot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1]) # square area
languages=['English','Urdu','German','Chinese','Arabic']
people=[100,50,150,200,70]
plt.xlabel("Language")
plt.ylabel("No of People")
ax.bar(languages,people,color="green")
plt.show()

```

Pie Chart - to find the distribution of the data

```python
figure1=plt.figure()
ax=figure1.add_axes([0,0,1,1])
languages=['English','Urdu','German','Chinese','Arabic']
people=[100,50,150,200,70]
plt.pie(people,labels=languages,autopct="%1.1f%%")
plt.show()
```

Scatter Plot


```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
figure2=plt.figure()
ax=figure2.add_axes([0,0,1,1])
ax.scatter(x,y,color="red")
ax.scatter(x,z,color="blue")
plt.show()
```

3D Plot

```python
figure3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.rand(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,c=z,cmap="Blues")
plt.show()
```
